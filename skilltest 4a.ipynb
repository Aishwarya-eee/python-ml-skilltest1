```python
import pandas as pd

# Create EEG data for participants
data = {
    "Participant": ["P1","P1","P2","P2"],
    "Electrode1": [5,7,6,8],
    "Electrode2": [3,4,5,6]
}

df = pd.DataFrame(data)

# Mean
print("Mean:")
print(df[["Electrode1","Electrode2"]].mean())

# Variance
print("Variance:")
print(df[["Electrode1","Electrode2"]].var())

# Peak values
print("Peak:")
print(df[["Electrode1","Electrode2"]].max())
```
