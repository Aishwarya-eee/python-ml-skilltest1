```python
import numpy as np


steps = np.array([
    8000, 9500, 10200, 12000, 7000, 8500, 11000,
    9000, 7600, 13000, 12500, 9800, 10500, 8700,
    
])


print("Total Steps:", np.sum(steps))
print("Average Steps per Day:", np.mean(steps))
print("Maximum Steps:", np.max(steps))
print("Minimum Steps:", np.min(steps))


print("\nHigh Activity Days (>10000 steps):")
for i in range(len(steps)):
    if steps[i] > 10000:
        print("Day", i+1, ":", steps[i], "steps")


print("\nTrend Analysis:")
for i in range(len(steps)-1):
    if steps[i+1] > steps[i]:
        print("Day", i+1, "to Day", i+2, ": Increase")
    elif steps[i+1] < steps[i]:
        print("Day", i+1, "to Day", i+2, ": Decrease")
    else:
        print("Day", i+1, "to Day", i+2, ": No Change")
```
