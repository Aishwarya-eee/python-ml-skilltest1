```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(1)
grades = np.random.randint(40, 100, size=(30, 3))

subjects = ["Math", "Science", "English"]

print("Student Grades (Math, Science, English):\n")
print(grades)


print("\nSubject-wise Statistics:")
for i in range(3):
    avg = np.mean(grades[:, i])
    highest = np.max(grades[:, i])
    lowest = np.min(grades[:, i])

    print(f"\n{subjects[i]}")
    print("Average:", avg)
    print("Highest:", highest)
    print("Lowest:", lowest)


total_scores = np.sum(grades, axis=1)
average_scores = np.mean(grades, axis=1)

print("\nTotal Scores of Students:\n", total_scores)
print("\nAverage Scores of Students:\n", average_scores)


top_student = np.argmax(total_scores)

print("\nTop Performing Student Index:", top_student)
print("Marks:", grades[top_student])
print("Total Score:", total_scores[top_student])
print("Average Score:", average_scores[top_student])


plt.figure(figsize=(10,5))

for i in range(3):
    plt.hist(grades[:, i], bins=8, alpha=0.5, label=subjects[i])

plt.title("Grade Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.legend()

plt.show()
```
